**1. Buat Gamma Correction**

```python
print(' Gamma Correction pada citra ')
print('----------------------------------')
try:
 gamma = int(input('Masukkan nilai Gamma: '))
except ValueError:
 print('Error, not a number')

img = cv.imread('/content/drive/MyDrive/PVCK/Images/female.tiff')
gamma_img = np.zeros(img.shape, img.dtype)

for y in range(img.shape[0]):
  for x in range(img.shape[1]):
    for c in range(img.shape[2]):
      gamma_img[y,x,c]= np.clip(255*pow(img[y,x,c]/255, (1/gamma)), 0, 255)

final_frame = cv.hconcat((img, gamma_img))
cv2_imshow(final_frame)
```

**2. Buat Simulas image depth**

```python
bit_depth=2
level = 255 / (pow(2, bit_depth)-1)
img = cv.imread('/content/drive/MyDrive/PVCK/Images/female.tiff', cv.IMREAD_GRAYSCALE)
depth_image = np.zeros(img.shape, img.dtype)

for y in range(img.shape[0]):
  for x in range(img.shape[1]):
    depth_image[y,x]=np.clip(round(img[y,x]/level)*level,0,255)

final_frame = cv.hconcat((img, depth_image))
plt.imshow(final_frame, cmap='gray')



```

**3. Buat modul Avarage Denoising**

```python
#import math
from math import log10, sqrt, pow
import os
import glob

def PSNR(img1, img2):
  mse = np.mean((img1 - img2)*2)
  if(mse==0): #tidak ada noise
    return 100
  max_pixel = 255.0
  psnr = 20*log10(max_pixel/sqrt(mse))
  return psnr

original = cv.imread('/content/drive/MyDrive/PVCK/Images/galaxy.jpg')
cv_img = []
for img in glob.glob('/content/drive/MyDrive/PVCK/Images/noises/*.jpg'):
  n = cv.imread(img)
  cv_img.append(n)

denoised_image = np.zeros(cv_img[0].shape)
jumlahGambar = 20
for i in range(jumlahGambar):
  denoised_image = denoised_image + cv_img[i]
denoised_image = np.uint16(denoised_image/jumlahGambar)

psnr = PSNR(original, denoised_image)

print('Nilai PSNRnya adalah: ', psnr, 'dB')
cv2_imshow(cv_img[0])
cv2_imshow(denoised_image)
```

**Nomor 5**

Menggunakan operator

* **OR**

```python
img_or = cv.imread('/content/drive/MyDrive/PVCK/Images/couple.tiff')
mask = np.zeros(img_or.shape, dtype=np.uint8)
mask = cv.circle(mask, (70, 50), 50, (255,255,255), -1)
result = img_or | mask

mask2= np.zeros(img_or.shape, dtype=np.uint8)
mask2= cv.circle(mask, (190, 50), 50, (255,255,255), -1)

result = img_or | mask2

final_frame = cv.hconcat((img_or, mask, result))
cv2_imshow(final_frame)
```

* **NOT**

```python
img_not = cv.imread('/content/drive/MyDrive/PVCK/Images/couple.tiff')
mask = np.zeros(img_not.shape, dtype=np.uint8)
mask = cv.circle(mask, (70, 50), 50, (255,255,255), -1)
result = ~img_not & ~mask

mask2= np.zeros(img_not.shape, dtype=np.uint8)
mask2= cv.circle(mask, (190, 50), 50, (255,255,255), -1)

result = ~img_not & ~mask2

final_frame = cv.hconcat((img_not, mask, result))
cv2_imshow(final_frame)
```

* **NAND**

```python
img_nand = cv.imread('/content/drive/MyDrive/PVCK/Images/couple.tiff')
mask = np.zeros(img_nand.shape, dtype=np.uint8)
mask = cv.circle(mask, (70, 50), 50, (255,255,255), -1)
result = ~img_nand & mask

mask2= np.zeros(img_nand.shape, dtype=np.uint8)
mask2= cv.circle(mask, (190, 50), 50, (255,255,255), -1)

result = ~img_nand & mask2

final_frame = cv.hconcat((img_nand, mask, result))
cv2_imshow(final_frame)
```

* **XOR**

```python
img_xor = cv.imread('/content/drive/MyDrive/PVCK/Images/couple.tiff')
mask = np.zeros(img_xor.shape, dtype=np.uint8)
mask = cv.circle(mask, (70, 50), 50, (255,255,255), -1)
result = img_xor ^ mask

mask2= np.zeros(img_xor.shape, dtype=np.uint8)
mask2= cv.circle(mask, (190, 50), 50, (255,255,255), -1)

result = img_xor ^ mask2

final_frame = cv.hconcat((img_xor, mask, result))
cv2_imshow(final_frame)
```

Kesimpulan:
1. OR --> menambahkan mask pada citar couple
2. NOT --> membuat citra menjadi negative
3. NAND --> menambahkan mask dan membuat objek citra dalam mask menjadi negative
4. XOR --> menambahkan mask dan membuat objek dalam mask menjadi negative, citra diluar mask akan tetap seperti gambar asli.
